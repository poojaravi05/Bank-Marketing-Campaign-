## Data Cleansing and Transformation

### Importing the required Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Reading the csv file

```python
bankData = pd.read_csv("bank_data.csv",delimiter=";")
bankData.head()
```

### Exploring the dataset

```python
bankData.info()
```

```python
bankData.describe().transpose()
```

### Checking for missing/null values

```python
bankData.isna().sum()
```

There are no null/missing values in the dataset.

### Checking the unique values in categorical columns

```python
catCols = ['job', 'marital', 'education', 'default', 'housing', 'loan','contact', 'month', 'day_of_week','poutcome', 'y']
numCols = ['age', 'duration', 'campaign', 'pdays','previous', 'emp.var.rate', 'cons.price.idx','cons.conf.idx', 'euribor3m']
```

```python
bankData.columns
```

```python
def findUnique(data):
    return data.unique()
```

```python
print("Unique Values: ")
for col in catCols:
    print(col,": ",findUnique(bankData[col]))
```

### Checking for Duplicate Data

```python
bankData.loc[bankData.duplicated() == True]
```

```python
bankData = bankData.drop_duplicates(keep=False)
```

```python
bankData.info()
```

Removed the duplicate rows from the dataset.

### Checking for Outliers

```python
for col in numCols:
    sns.boxplot(data=bankData,x=col)
    plt.show()
```

```python
#Duration
bankData.loc[bankData["duration"]>4000]
```

```python
#Campaign
bankData.loc[bankData["campaign"]>50]
```

```python
#Cons.conf.idx
bankData.loc[bankData["cons.conf.idx"]>-30]
```

```python
#Removing the rows [4107, 22192, 24091]
rowsToRemove = [4107, 22192, 24091]
bankData = bankData.drop(rowsToRemove,axis=0)
```

```python
#Box plots after removing the rows with outliers
for col in numCols:
    sns.boxplot(data=bankData,x=col)
    plt.show()
```

### Checking PDays

```python
bankData["pdays"].value_counts()
```

```python
bankData["poutcome"].value_counts()
```

```python
bankData.loc[(bankData["pdays"]==999) & (bankData["poutcome"]=='failure')]
```

```python
#Converting the value of poutcome for rows with value 999 in pdays
bankData.loc[bankData["pdays"]==999,"poutcome"] = "nonexistent"
bankData.loc[(bankData["pdays"]==999) & (bankData["poutcome"]=='failure')]
```

```python
bankData.loc[(bankData["pdays"]==999),"poutcome"].value_counts()
```

### Final Cleaned Dataset

```python
bankData.info()
```

### Performing EDA

```python
#The number of 'yes' and 'no' values in the response column 'y'.

value_counts = bankData['y'].value_counts()

plt.bar(value_counts.index, value_counts.values)
plt.title('Yes Or No Responses')
plt.show()
```

### Comparing Term deposit(y) with other columns

```python
#age and y

age_bankData = pd.DataFrame()

age_bankData['age_yes'] = bankData[bankData['y'] == 'yes']['age'].describe()
age_bankData['age_no'] = bankData[bankData['y'] == 'no']['age'].describe()

age_bankData = age_bankData.loc[['count', 'mean', 'std', 'min', '25%', '50%', '75%', 'max']]

age_bankData
```

```python
age_bankData.drop(['count', '25%', '50%', '75%']).plot(kind='bar', title='Age and Term Deposit')
plt.show()
```

```python
#job and y

job_bankData = pd.DataFrame()

job_bankData['yes'] = bankData[bankData['y'] == 'yes']['job'].value_counts()
job_bankData['no'] = bankData[bankData['y'] == 'no']['job'].value_counts()

fig, ax = plt.subplots()
job_bankData.plot(kind='bar', ax=ax)
ax.set_title('Job and Term Deposit')

plt.show()
```

```python
#marital status and y

marital_bankData = pd.DataFrame()

marital_bankData['yes'] = bankData[bankData['y'] == 'yes']['marital'].value_counts()
marital_bankData['no'] = bankData[bankData['y'] == 'no']['marital'].value_counts()

fig, ax = plt.subplots()
marital_bankData.plot(kind='bar', ax=ax)
ax.set_title('Marital Status and Term Deposit')

plt.show()
```

```python
#education and y

education_bankData = pd.DataFrame()

education_bankData['yes'] = bankData[bankData['y'] == 'yes']['education'].value_counts()
education_bankData['no'] = bankData[bankData['y'] == 'no']['education'].value_counts()

fig, ax = plt.subplots()
education_bankData.plot(kind='bar', ax=ax)
ax.set_title('Education and Term Deposit')

plt.show()
```

```python
#Contact and y

contact_bankData = pd.DataFrame()

contact_bankData['yes'] = bankData[bankData['y'] == 'yes']['contact'].value_counts()
contact_bankData['no'] = bankData[bankData['y'] == 'no']['contact'].value_counts()

fig, ax = plt.subplots()
contact_bankData.plot(kind='bar', ax=ax)
ax.set_title('Contact and Deposit')

plt.show()
```

```python
#campaign and y

campaign_bankData = pd.DataFrame()

campaign_bankData['campaign_yes'] = bankData[bankData['y'] == 'yes']['age'].describe()
campaign_bankData['campaign_no'] = bankData[bankData['y'] == 'no']['age'].describe()

campaign_bankData = campaign_bankData.loc[['count', 'mean', 'std', 'min', '25%', '50%', '75%', 'max']]

campaign_bankData
```

```python
campaign_bankData.drop(['count', '25%', '50%', '75%']).plot(kind='bar', title='Campaign and Term Deposit')
plt.show()
```
